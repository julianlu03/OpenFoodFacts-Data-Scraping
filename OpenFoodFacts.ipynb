```python
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import requests 

def retrieve_info(url):
    try:
        driver = webdriver.Chrome()
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html')
        driver.get(url)
        table = soup.find_all("table")[0]
        rows = table.find_all("tr")
        rows = [row.text.replace("\n", "") for row in rows]
        rows = [row.strip(" ") for row in rows]
        nutrition_facts = []
        
        for row in rows:    
            element = row.split("  ")
            for i in range(len(element)):
                if element[i] != '': #Removes all empty strings in the list
                    nutrition_facts.append(element[i])
                    
        #Returns name
        def get_name():
            try:
                if driver.find_element(By.CLASS_NAME, 'title-1').text.strip():
                    name = driver.find_element(By.CLASS_NAME, 'title-1').text.strip()
                    return name
            except:
                return None
        food_name = get_name()
        #Returns brand name
        def get_brand():
            try:
                if driver.find_element(By.ID, 'field_brands_value').text.strip():
                    brand = driver.find_element(By.ID, 'field_brands_value').text.strip()
                    return brand
            except:
                return None
        
        brand_name = get_brand()
        #Returns Category
        def get_category():
            try:
                if driver.find_element(By.ID, 'field_categories_value').text.strip():
                    category = driver.find_element(By.ID, 'field_categories_value').text.strip()
                    return category
            except:
                return None
        
        category = get_category()
        #Returns awards
        def get_awards():
            try:
                if driver.find_element(By.ID, 'field_labels_value').text.strip():
                    awards = driver.find_element(By.ID, 'field_labels_value').text.strip()
                    return awards
            except:
                return None
    
        awards = get_awards()
        #Returns barcode
        def get_barcode():
            try:
                if driver.find_element(By.ID, 'barcode_paragraph').text.strip():
                    barcode = ((driver.find_element(By.ID, 'barcode_paragraph').text.strip()).split(":")[1]).strip()
                    return barcode
            except:
                return None
        
        barcode = get_barcode()
        #Returns ingredient count
        def get_ingredient_count():
            try:
                if driver.find_element(By.XPATH, '//*[@id="panel_ingredients"]/li/a/h4').text.strip():
                    ingredient_count = driver.find_element(By.XPATH, '//*[@id="panel_ingredients"]/li/a/h4').text.strip()
                    if ingredient_count == "Ingredients are missing":
                        return None
                    return ingredient_count
            except:
                return None
        ingredients = get_ingredient_count()
        if ingredients == None:
            pass
        else:
            ingredients = ingredients.split(" ")
            ingredients = ingredients[0]
       #Returns serving size 
        def get_serving_size():
            try:
                if driver.find_element(By.XPATH, '//*[@id="panel_serving_size_content"]/div/div/div').text.strip():
                    serving_size = serving_size = driver.find_element(By.XPATH, '//*[@id="panel_serving_size_content"]/div/div/div').text.strip()
                    serving_size = (serving_size.split(':')[1]).strip()
                    return serving_size
            except:
                return None
        serving_size = get_serving_size()
        #Returns Nutri-Score
        def get_nutri_score():
            try:
                if driver.find_element(By.XPATH, '//*[@id="attributes_grid"]/li[1]/a/div/div/div[2]/h4').text.strip():
                    nutri_score = driver.find_element(By.XPATH, '//*[@id="attributes_grid"]/li[1]/a/div/div/div[2]/h4').text.strip()
                    if nutri_score == "Nutri-Score unknown":
                        return None
                    nutri_score = nutri_score[-1]
                    return nutri_score
            except:
                return None
        nutri_score = get_nutri_score()
        if nutri_score == None:
            pass
        else:
            nutri_score = nutri_score[-1]
        #Returns Nova Score
        def get_nova_score():
            try:
                if driver.find_element(By.XPATH, '//*[@id="attributes_grid"]/li[2]/a/div/div/div[2]/h4').text.strip():
                    nova_score = driver.find_element(By.XPATH, '//*[@id="attributes_grid"]/li[2]/a/div/div/div[2]/h4').text.strip()
                    if nova_score == "NOVA not computed":
                        return None
                    return nova_score
            except:
                return None
        nova_score = get_nova_score()
        if nova_score == None:
            pass
        else:
            nova_score = nova_score.split(" ")
            nova_score = nova_score[-1]
        #Returns dictionary with all above info and rest of nutrition info 
        nutrition_dict = {
            "Brand name": {food_name: brand_name},
            "Category": {food_name: category},
            "Awards": {food_name: awards},
            "Barcode": {food_name: barcode},
            "Number of ingredients": {food_name: ingredients},
            "Serving size": {food_name: serving_size},
            "Nutri-Score": {food_name: nutri_score},
            "Nova Score": {food_name: nova_score},
            "Energy (per standard)": {food_name: None},
            "Energy (per serving)": {food_name: None},
            "Energy from fat (per standard)": {food_name: None},
            "Energy from fat (per serving)": {food_name: None},
            "Fat (per standard)": {food_name: None},
            "Fat (per serving)": {food_name: None},
            "Saturated fat (per standard)": {food_name: None},
            "Saturated fat (per serving)": {food_name: None},
            "Trans fat (per standard)": {food_name: None},
            "Trans fat (per serving)": {food_name: None},
            "Cholesterol (per standard)": {food_name: None},
            "Cholesterol (per serving)": {food_name: None},
            "Carbohydrates (per standard)": {food_name: None},
            "Carbohydrates (per serving)": {food_name: None},
            "Sugars (per standard)": {food_name: None},
            "Sugars (per serving)": {food_name: None},
            "Fiber (per standard)": {food_name: None},
            "Fiber (per serving)": {food_name: None},
            "Proteins (per standard)": {food_name: None},
            "Proteins (per serving)": {food_name: None},
            "Salt (per standard)": {food_name: None},
            "Salt (per serving)": {food_name: None},
            "Vitamin A (per standard)": {food_name: None},
            "Vitamin A (per serving)": {food_name: None},
            "Vitamin C (ascorbic acid) (per standard)": {food_name: None},
            "Vitamin C (ascorbic acid) (per serving)": {food_name: None},
            "Calcium (per standard)": {food_name: None},
            "Calcium (per serving)": {food_name: None},
            "Iron (per standard)": {food_name: None},
            "Iron (per serving)": {food_name: None},
            "Fruits‚ vegetables‚ nuts and rapeseed‚ walnut and olive oils (estimate from ingredients list analysis) (per standard)": {food_name: None},
            "Fruits‚ vegetables‚ nuts and rapeseed‚ walnut and olive oils (estimate from ingredients list analysis) (per serving)": {food_name: None}    
        }
        nutrition_keys = {
            "Energy": "Energy",
            "Energy from fat": "Energy from fat",
            "Fat": "Fat",
            "Saturated fat": "Saturated fat",
            "Trans fat": "Trans fat",
            "Cholesterol": "Cholesterol",
            "Carbohydrates": "Carbohydrates",
            "Sugars": "Sugars",
            "Fiber": "Fiber",
            "Proteins": "Proteins",
            "Salt": "Salt",
            "Vitamin A": "Vitamin A",
            "Vitamin C (ascorbic acid)": "Vitamin C (ascorbic acid)",
            "Calcium": "Calcium",
            "Iron": "Iron",
            "Fruits‚ vegetables‚ nuts and rapeseed‚ walnut and olive oils (estimate from ingredients list analysis)": "Fruits‚ vegetables‚ nuts and rapeseed‚ walnut and olive oils (estimate from ingredients list analysis)"
        }
        for i in range(len(nutrition_facts)):
            nutrient = nutrition_facts[i]           
            if nutrient in nutrition_keys:
                if nutrition_facts[2][:18] != "As soldper serving":
                    key_standard = f"{nutrition_keys[nutrient]} (per standard)"
                    nutrition_dict[key_standard][food_name] = nutrition_facts[i + 1]
                else:
                    if nutrient == "Energy" or nutrient == "Energy from fat":
                        key_standard = f"{nutrition_keys[nutrient]} (per standard)"
                        key_serving = f"{nutrition_keys[nutrient]} (per serving)"
                        nutrition_dict[key_standard][food_name] = nutrition_facts[i + 1]
                        try:
                            nutrition_dict[key_serving][food_name] = nutrition_facts[i + 2]
                        except IndexError:
                            pass
                    else:
                        key_standard = f"{nutrition_keys[nutrient]} (per standard)"
                        key_serving = f"{nutrition_keys[nutrient]} (per serving)"
                        per_standard_value = (nutrition_facts[i + 1]).split(" ")
                        per_standard_value = per_standard_value[0]
                        nutrition_dict[key_standard][food_name] = per_standard_value
                        try:
                            per_serving_value = (nutrition_facts[i + 2]).split(" ")
                            per_serving_value = per_serving_value[0]
                            nutrition_dict[key_serving][food_name] = per_serving_value
                        except IndexError:
                            pass
        global df
        df = pd.DataFrame(nutrition_dict)
    except:
        print(f"Error occured at Row {row_number}.")

have_header = True
for link in url_list:
    row_number = url_list.index(link) + 1
    retrieve_info(link)
    df.to_csv(r"C:\Users\julia\OneDrive\Documents\Web Scraping\Folder for Output\C_StoreOutput.csv", mode = 'a', header = have_header)
    have_header = False
    sleep(randint(5, 12))
```
